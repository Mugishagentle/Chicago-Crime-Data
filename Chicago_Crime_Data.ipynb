```python
# Import required packages to load the data
import pandas as pd

# Save the filepath and load the data as "df"
fpath = "/content/drive/MyDrive/CodingDojo/04-DataViz/Week13/Data/Chicago-Crime_2022.csv"
df = pd.read_csv(fpath)
```

#Topic 1: Comparing Police Districts
### Which district had the most crimes in 2022?


```python
# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d %H:%M:%S')


# Filter data for the year 2022
df_2022 = df[df['Date'].dt.year == 2022]

# Calculate total crimes in each district for 2022
district_crime_counts_2022 = df_2022['District'].value_counts()

# District with the most crimes
district_most_crimes_2022 = district_crime_counts_2022.idxmax()
most_crimes_count_2022 = district_crime_counts_2022.max()

# District with the least crimes
district_least_crimes_2022 = district_crime_counts_2022.idxmin()
least_crimes_count_2022 = district_crime_counts_2022.min()

print(f"District {district_most_crimes_2022} had the most crimes in 2022 with {most_crimes_count_2022} crimes.")
print(f"District {district_least_crimes_2022} had the least crimes in 2022 with {least_crimes_count_2022} crimes.")
```

# Topic 3: Comparing AM vs. PM Rush Hour
## Are crimes more common during AM rush hour or PM rush hour?

```python
# Filter data for AM rush hour (7 AM - 10 AM)
am_rush_hour = df[(df['Date'].dt.hour >= 7) & (df['Date'].dt.hour < 10)]

# Filter data for PM rush hour (4 PM - 7 PM)
pm_rush_hour = df[(df['Date'].dt.hour >= 16) & (df['Date'].dt.hour < 19)]

# Analyze the distribution of crimes during AM and PM rush hours
total_crimes_am = len(am_rush_hour)
total_crimes_pm = len(pm_rush_hour)

print("Total crimes during AM rush hour:", total_crimes_am)
print("Total crimes during PM rush hour:", total_crimes_pm)

# Identify the top 5 most common crimes during AM and PM rush hours
top_5_am_crimes = am_rush_hour['Primary Type'].value_counts().head(5)
top_5_pm_crimes = pm_rush_hour['Primary Type'].value_counts().head(5)

print("\nTop 5 most common crimes during AM rush hour:")
print(top_5_am_crimes)

print("\nTop 5 most common crimes during PM rush hour:")
print(top_5_pm_crimes)

# Determine the frequency of Motor Vehicle Thefts during AM and PM rush hours
mv_thefts_am = len(am_rush_hour[am_rush_hour['Primary Type'] == 'MOTOR VEHICLE THEFT'])
mv_thefts_pm = len(pm_rush_hour[pm_rush_hour['Primary Type'] == 'MOTOR VEHICLE THEFT'])

print("\nFrequency of Motor Vehicle Thefts during AM rush hour:", mv_thefts_am)
print("Frequency of Motor Vehicle Thefts during PM rush hour:", mv_thefts_pm)

# Determine if Motor Vehicle Thefts are more common during AM or PM rush hour
if mv_thefts_am > mv_thefts_pm:
    print("Motor Vehicle Thefts are more common during AM rush hour.")
elif mv_thefts_am < mv_thefts_pm:
    print("Motor Vehicle Thefts are more common during PM rush hour.")
else:
    print("Motor Vehicle Thefts have the same frequency during both AM and PM rush hours.")
```

# Topic 4: Comparing Months
## What months have the most crime? Least?

```python
# Extract month from the 'Date' column
df['Month'] = df['Date'].dt.month

# Calculate total number of crimes per month
crimes_per_month = df.groupby('Month').size()

# Find the month with the most crime and the month with the least crime
month_most_crime = crimes_per_month.idxmax()
month_least_crime = crimes_per_month.idxmin()

print("Month with the most crime:", month_most_crime)
print("Month with the least crime:", month_least_crime)

# Identify individual crimes whose frequency doesn't follow the overall pattern of crime by month
# Compare the frequency of each crime to the average frequency across all months
average_crime_per_month = crimes_per_month.mean()
individual_crimes = df['Primary Type'].unique()

print("\nIndividual crimes with frequency differing from the overall pattern by month:")
for crime in individual_crimes:
    crime_freq_per_month = df[df['Primary Type'] == crime].groupby('Month').size()
    if abs(crime_freq_per_month.mean() - average_crime_per_month) > 100:  # Adjust threshold as needed
        print(f"{crime}: {crime_freq_per_month.mean()}")
```
